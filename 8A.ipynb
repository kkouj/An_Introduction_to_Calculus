```python
from sympy import *
import numpy as np

a,b,c, x, y, z, t = symbols('a b c x y z t')
k, m, n = symbols('k m n', integer=True)
f, g, h = symbols('f g h', cls=Function)
init_printing()
```

```python
# 分母分子別々に定義する
f = (1 - cos(6 * x)) 
g = ( 1 - cos(3 * x))
```

```python
# 下式のx --> 0を調べる。このままだと0/0の不定形
h = f / g
h
```

```python
# 寄り道：グラフを書くと、x=0のとき4あたりにありそう
plot(h, (x,-3, 3))
```

```python
# 寄り道：そのままsympyのlimit関数を適用すると4
limit(h, x, 0)
```

```python
# ロピタルの定理を使う。でもその結果も0/0の不定形
diff(f) / diff(g)
```

```python
#まだゼロの不定形なのでもう一度ロピタルの定理を使う。その結果、x --> 0の極限は4になることがわかる
diff(diff(f)) / diff(diff(g))
```
